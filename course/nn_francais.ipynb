## Section convert txt to csv

```python
def format_text(path):
    with open(path, "r") as fich:
        text = fich.read()
        text = text.split('\n', 1)[1]
        text = text.replace("\n", " ")
        text = text.replace("</doc>", "")
        return text
```

```python
df = pd.DataFrame(columns=["text"])
list_of_articles = os.listdir(dest)
num_of_articles = len(list_of_articles)
```

```python
list_of_articles[:10]
```

```python
num_of_articles
```

```python
for article in tqdm(list_of_articles):
    text = format_text(f"{dest}/{article}")
    tdf = pd.DataFrame([text], columns=["text"])
    df = df.append(tdf)
df = df.reset_index(drop=True)
df.head()
```

```python
tokenizer = Tokenizer(lang='fr', n_cpus=24)
```

```python
data_lm_full = (TextList.from_csv(path, csv_name='50ktrain.csv', cols=0, processor=[TokenizeProcessor(tokenizer=tokenizer), NumericalizeProcessor(max_vocab=60000)]))

                
```

```python
data_lm_full = data_lm_full.split_from_df(col=1)
```

```python
data_lm_full = data_lm_full.label_for_lm()
```

```python
data_lm_full = data_lm_full.databunch(bs=bs, num_workers=24)
```
